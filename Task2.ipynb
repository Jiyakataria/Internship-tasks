```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Manually given revenue values
weekday = [420, 480, 500, 550, 610, 470, 530]
weekend = [520, 540, 560, 580, 600, 590, 570]

df = pd.DataFrame({
    "Day": ["Weekday"]*len(weekday) + ["Weekend"]*len(weekend),
    "Bill": weekday + weekend
})

# Box Plot
sns.boxplot(x="Day", y="Bill", data=df)
plt.title("Weekday vs Weekend Revenue Stability")
plt.show()

# Histogram
sns.histplot(data=df, x="Bill", hue="Day")
plt.title("Bill Distribution")
plt.show()

# Line Graph (Average Bill)
avg = df.groupby("Day")["Bill"].mean()
plt.plot(avg.index, avg.values, marker="o")
plt.title("Average Bill Trend")
plt.xlabel("Day")
plt.ylabel("Average Bill")
plt.show()
```
